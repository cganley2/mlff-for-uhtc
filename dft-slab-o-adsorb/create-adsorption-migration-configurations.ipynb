```python
slab144 = ase.io.read('/home/cganley2/scitech2026/pre-training-data/zrb2-slab-001/qe-calcs/relax-144atoms/zrb2-slab-001-relax.out')
add_adsorbate(slab144, 'O', 1.5, position=(4.77,4.59))
write_espresso_in('zrb2-o-T-relax.in', slab144, input_data=input_data, pseudopotentials=pseudopotentials, kpts=kpts)
```
